```python
def fix_price(x):
    if type(x) == float:
        return x
    else:
        if x[1] == 'Lac':
            return round(float(x[0])/100,2)
        else:
            return round(float(x[0]),2)

        
    
```

```python
df = df.loc[df['price'] != 'Price on Request']
```

```python
df.loc[:, 'price'] = df['price'].str.split(' ').apply(fix_price)
```
